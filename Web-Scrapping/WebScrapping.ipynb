```python
!pip install pandas
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://editorial.rottentomatoes.com/guide/best-movies-of-all-time/'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
print(soup)
```

```python
movie_name = []
year = []
rating = []
critics_consensus = []
synopsis = []
director_list = [] #list for director
```

```python
movie_data = soup.findAll('div', attrs= {'class': 'row countdown-item'})
```

```python
movie_data
```

```python
for store in movie_data:  # Correct: for loop header

    name = store.h2.a.text  # Correct: indented
    movie_name.append(name)  # Correct: indented

    year_of_release = store.h2.find('span', class_='subtle start-year').text.replace('(', '').replace(')', '')  # Correct: indented
    year.append(year_of_release)  # Correct: indented

    rate = store.h2.find('img', class_='icon tiny').span.text  # Correct: indented
    rating.append(rate)  # Correct: indented

    critics = store.find('div', class_='info critics-consensus').text.replace('\n', '')  # Correct: indented
    critics_consensus.append(critics)  # Correct: indented

    requirment = store.find('div', class_='info synopsis').text.replace('\n', '')  # Correct: indented
    synopsis.append(requirment)  # Correct: indented

    directer = store.find('div', class_='info director').a.text  # Correct: indented
    director_list.append(directer) #Correct: indented
    
    # ... (rest of your code)
```

```python
movie_DF = pd.DataFrame({'Name of movie': movie_name, 
                         'Year of relase': year,
                          'Movie Rating': rating,
                          'Critics': critics_consensus,
                          'synopsis': synopsis,
                         "Director": director_list})
```

```python
#Saving data in Excel file:
!pip install openpyxl
movie_DF.to_excel("Top_100_IMDB_Movies.xlsx")
```
